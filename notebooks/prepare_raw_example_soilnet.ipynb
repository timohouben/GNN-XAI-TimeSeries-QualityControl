```python
import xarray as xr
import glob, os
import numpy as np
import pandas as pd
```

```python
# load raw dataset
raw_ds_path = '/bg/home/lasota-e/cleaning/experiment_pipeline/data/soilnet/proc_soilnet_prepreproc.nc'
raw_ds = xr.open_dataset(raw_ds_path)
```

```python
start_date = '2014-08-01 00:00'
end_date = '2014-10-29 00:00'
start_date_pd = pd.to_datetime(start_date)
end_date_pd = pd.to_datetime(end_date)
#end_date_pd = pd.to_datetime(start_date) + np.timedelta64(interval2[i], 'h')
```

```python
ds_daterange = raw_ds.sel(time=slice(start_date_pd, end_date_pd))
ind_drop = np.where(np.isnan(ds_daterange.moisture.values).all(axis=1))
ds_daterange = ds_daterange.drop_isel(sensor_id=ind_drop)
```
